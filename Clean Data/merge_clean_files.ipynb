# Merging Files and Extracting Sector Information

- During the data merging process, I identified that the sector information was embedded within the `society` column. To extract this, I split the contents of the `society` field and parsed the relevant portion corresponding to the sector.

## Challenge:

- Upon reviewing the extracted data, I discovered that in many instances, area or locality names were listed instead of actual sector numbers. As a result, automated extraction was not feasible in all cases. The only viable approach for resolving these entries is to manually research the areas (e.g., via Google) to accurately determine their corresponding sector numbers.
